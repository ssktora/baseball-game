# 野球ゲーム
投手は4つのコースのいずれかに投げる。打者は4つのコースのいずれかをスイングする。投手が投げたコースと打者が振ったコースが一致したら5%の確率でホームラン、5%の確率でスリーベース、5%の確率でアウト、70%の確率でヒット、15%の確率でツーベースとする。一致しなかったときはストライクとなり、３ストライクになると三振でアウトになる。

## 一球の勝負の関数を生成

```python
import random
course = {"0":"内高め","1":"内低め","2":"外高め","3":"外低め"}
def match():
    player = str(input("数字を入力"))
    print("あなたは" + course[player] + "を選びました。")
    com = str(random.randrange(4))
    print("対戦相手は" + course[com] + "を選びました。")
    if com == player:
        roulette = random.randrange(100)
        if roulette in [x for x in range(5)]:
            print("ホームラン！！")
            return 4
        elif roulette in [x for x in range(5, 10)]:
            print("スリーベースヒット")
            return 3
        elif roulette in [x for x in range(10, 15)]:
            print("アウト")
            return "out"
        elif roulette in [x for x in range(15, 85)]:
            print("ヒット")
            return 1
        else:
            print("ツーベースヒット")
            return 2
    else:
        print("ストライク")
        return "strike"
        
```

## １打席の勝負の関数を生成

```python
def at_bat():
    strike_count = 0
    while strike_count < 3:
        one_ball = match()
        if one_ball == "strike":
            strike_count += 1
            print("ストライクカウント", strike_count)
        else:
            return one_ball
    print("三振")
    return "out"


            
```

# 走者に関する関数を生成

```python
def pointer(num, a_list, point):
    for i in range(len(a_list)):
        a_list[i] = a_list[i] + num
    a_list.append(num)
    point_list  = [1 for x in a_list if x>=4]
    if len(point_list) >= 1:
        point += len(point_list)
        for i in range(len(point_list)):
            a_list.pop(0)
        print("この回の得点", point)
        point_list = []
    return a_list, point
        
def runner(a_list, point, result, count):
    if result == 4:
        a_list, point = pointer(4, a_list, point)
        print("ランナー", a_list)
    elif result == 3:
        a_list, point = pointer(3, a_list, point)
        print("ランナー", a_list)
    elif result == 2:
        a_list, point = pointer(2, a_list, point)
        print("ランナー", a_list)
    elif result == 1:
        a_list, point = pointer(1, a_list, point)
        print("ランナー", a_list)
    else:
        count += 1
        print("アウトカウント", count)
    return a_list, point, count
        
        
    
```

# 1イニングの勝負の関数を生成

```python
def inning_match():
    out_count = 0
    inning_point = 0
    runner_list = []
    while out_count != 3:
        result = at_bat()
        runner_list, inning_point, out_count = runner(runner_list, inning_point, result, out_count)
    print("この回の得点", inning_point)
    return inning_point
```

# 1試合の勝負の関数を生成

```python
def game():
    inning_count = 1
    player1_point = 0
    player2_point = 0
    while inning_count <= 8:
        print(str(inning_count) + "回表")
        player1_inning_point = inning_match()
        player1_point += player1_inning_point
        print("あなたの総得点：" + str(player1_point) + "点 VS 対戦相手の総得点：" + str(player2_point) + "点")
        print(str(inning_count) + "回裏")
        player2_inning_point = inning_match()
        player2_point += player2_inning_point
        print("あなたの総得点：" + str(player1_point) + "点 VS 対戦相手の総得点：" + str(player2_point) + "点")
        inning_count += 1
    
    print("9回表")
    player1_inning_point = inning_match()
    player1_point += player1_inning_point
    print("あなたの総得点：" + str(player1_point) + "点 VS 対戦相手の総得点：" + str(player2_point) + "点")
    
    if player1_point > player2_point:
        print("9回裏")
        player2_inning_point = inning_match()
        player2_point += player2_inning_point
        print("あなたの総得点：" + str(player1_point) + "点 VS 対戦相手の総得点：" + str(player2_point) + "点")
    
    print("試合結果")
    if player1_point > player2_point:
        print("あなたの勝ち")
    elif player1_point < player2_point:
        print("対戦相手の勝ち")
    else:
        print("引き分け")
    
    return player1_point, player2_point

game()
    
        
    
    
        
```
