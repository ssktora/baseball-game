```python
# プレイヤーの画像を読み込む
def load_player(image_filename):
    img = Image.open(image_filename)
    img_tk = ImageTk.PhotoImage(img)
    return img_tk
```

### 描画系


```python
# 迷路を表示する関数
def draw_map():
    # 左上から右下へと描画
    # 迷路の行数
    rows = len(map_data)
    # 迷路の列数
    cols = len(map_data[0])
    for y in range(rows):
        y1 = y * tile_size
        y2 = y1 + tile_size
        for x in range(cols):
            x1 = x * tile_size
            x2 = x1 + tile_size
            # 該当場所の値を得る
            p = map_data[y][x]
            # 値に応じた色を決定する
            if p == 0: color = "white"
            if p == 1: color = "#404040"
            if p == 2: color = "red"
            if p == 3: color = "blue"
            if p == 4: color = "green"
            # 正方形を描画
            canvas.create_rectangle(
                x1, y1, x2, y2, # 座標
                fill=color, # 塗色
                outline="black", width=2) # 枠線
```

```python
# プレイヤーを描画
def draw_player():
    x = px * tile_size
    y = py * tile_size
    canvas.create_image(x, y, image=player_image, anchor=NW)
```

### アイテム関連

```python
items = ["ハンマー", "ゴール変更機", "テレポートスティック"]
import random

def get_number():
    num = random.randint(0,100)
    if 0 <= num <= 49:
        return 0
    elif 50 <= num <= 74:
        return 1
    else:
        return 2
    

def up():
    map_data[py-1][px] = 0

def down():
    map_data[py+1][px] = 0

def left():
    map_data[py][px-1] = 0
    
def right():
    map_data[py][px+1] = 0
    
def hammer():
    tki = tkinter.Tk()
    tki.geometry('300x200')
    tki.title('破壊する壁の方向')
    # ボタン作成
    btn = tkinter.Button(tki, text='上', command=lambda:[up(), tki.destroy()])
    btn.place(x=140, y=170)
    btn = tkinter.Button(tki, text='下', command=lambda:[down(), tki.destroy()])
    btn.place(x=140, y=140)
    btn = tkinter.Button(tki, text='左', command=lambda:[left(), tki.destroy()])
    btn.place(x=140, y=110)
    btn = tkinter.Button(tki, text='右', command=lambda:[right(), tki.destroy()])
    btn.place(x=140, y=80)
    canvas.delete("all")
    draw_map()
    draw_player()
    items_now.remove("ハンマー")

def goal_change():
    checker = True
    while checker:
        x = random.randint(0,19)
        y = random.randint(0,25)
        if map_data[x][y] == 0:
            checker = False
    for i in range(20):
        for j in range(26):
            if map_data[i][j] == 3:
                gx = i
                gy = j
    map_data[x][y] = 3
    map_data[gx][gy] = 0
    canvas.delete("all")
    draw_map()
    draw_player()
    items_now.remove("ゴール変更機")
    
def tele():
    global px, py
    for i in range(26):
        for j in range(20):
            if map_data[j][i] == 3:
                gx = i
                gy = j
    if gx - 5 <= 0:
        down_x = 0
    else:
        down_x = gx - 5
    if gx + 5 >= 25:
        up_x = 25
    else:
        up_x = gx + 5
    if gy - 5 <= 0:
        down_y = 0
    else:
        down_y = gy - 5
    if gy + 5 >= 19:
        up_y = 19
    else:
        up_y = gy + 5
    checker = True
    while checker:
        x = random.randint(down_x,up_x)
        y = random.randint(down_y,up_y)
        if map_data[y][x] == 0:
            checker = False
    px = x
    py = y
    canvas.delete("all")
    draw_map()
    draw_player()
    items_now.remove("テレポートスティック")
            
    


```

### イベント系


```python
# キャンバスを左クリックした時の処理
items_now = []
import copy
def canvas_left_click(e):
    global items_now
    global px, py
    # マウス座標を得る
    # マウスのX座標
    mx = e.x
    # マウスのY座標
    my = e.y
    
    # 移動前に前回の値を覚えておく
    px_tmp = px
    py_tmp = py

    # プレイヤーが上下左右のどちらに動くか判定
    # プレイヤーの絶対座標を計算
    xx, yy = [px * tile_size, py * tile_size]
    ix = mx - xx
    iy = my - yy
    if abs(ix) > abs(iy):
        # 左右移動
        if ix > 0:
            px += 1
        else:
            px -= 1
    else:
        # 上下移動
        if iy > 0:
            py += 1
        else:
            py -= 1
    # 移動先がマップデータ外なら戻す
    global map_data
    if px < 0 or px >= len(map_data[0]):
        px = px_tmp
    if py < 0 or py >= len(map_data):
        py = py_tmp
    # 移動先が壁なら元の位置に戻す 
    mv = map_data[py][px]
    if mv == 1:
        px = px_tmp
        py = py_tmp
        msgbox.showinfo(message="壁にぶつかった")
        return
    # マップとプレイヤーを描画する
    canvas.delete("all")
    draw_map()
    draw_player()
    
    print("player={0},{1}".format(px, py))
    # ゴールにたどり着いたか？
    global n
    if mv == 3:
        msgbox.showinfo(message="祝！ゴール！")
        px = 2
        py = 2
        if n == 0:
            n = 1
            map_data = copy.deepcopy(all_map_data[n])
        elif n == 1:
            n = 2
            map_data = copy.deepcopy(all_map_data[n])
        else:
            n = 0
            map_data = copy.deepcopy(all_map_data[n])
        canvas.delete("all")
        draw_map()
        draw_player()
        return 
    #アイテムマスに到達したとき
    if mv == 4:
        item_num = get_number()
        items_now.append(items[item_num])
        msgbox.showinfo(message=items[item_num]+"を手に入れた")
        print(items_now)
        map_data[py][px] = 0
         
# キャンバスを右クリックした時の処理
def canvas_right_click(e):
    tki = tkinter.Tk()
    tki.geometry('300x200')
    tki.title('アイテム一覧')
    # ボタン作成
    x_p = 140
    y_p = 20
    for i in range(len(items_now)):
        if items_now[i] == "ハンマー":
            btn = tkinter.Button(tki, text=str(items_now[i]), command=lambda:[tki.destroy(), hammer()])
            btn.place(x=x_p, y=y_p)
            y_p += 30
        
        elif items_now[i] == "ゴール変更機":
            btn = tkinter.Button(tki, text=str(items_now[i]), command=lambda:[tki.destroy(), goal_change()])
            btn.place(x=x_p, y=y_p)
            y_p += 30
        
        elif items_now[i] == "テレポートスティック":
            btn = tkinter.Button(tki, text=str(items_now[i]), command=lambda:[tki.destroy(), tele()])
            btn.place(x=x_p, y=y_p)
            y_p += 30
        

        
        
    









```

### ウインドウとキャンバスの作成≒メイン


```python
# ウィンドウとキャンバスを作成
def create_window():
    # ウインドウを作成
    global win
    win = Tk()
    win.title("迷路")
    

    # 迷路データを読み込む
    global map_data
    global n
    n = 0
    map_data = copy.deepcopy(all_map_data[n])

    # 迷路の行数
    rows = len(map_data)
    # 迷路の列数
    cols = len(map_data[0])

    # キャンバスの作成
    global canvas
    canvas = Canvas(
        win,
        width=(cols * tile_size),
        height=(rows * tile_size))
    
    canvas.pack()
    

    # プレイヤー画像を読み込む
    global player_image
    player_image = load_player(player_image_file)

    # マップとプレイヤーを描画する
    draw_map()
    draw_player()
    # イベントを追加
    canvas.bind("<1>", canvas_left_click)
    canvas.bind("<Button-3>", canvas_right_click)
    

    #メインループ開始
    win.mainloop()
```

## メインプログラムの実行


```python
# ウィンドウを作成
create_window()
```
