```python
import pandas as pd
```

```python
df = pd.read_csv('チーム成績.csv', encoding='shift_jis')
# 不要な行と列を削除
df = df.dropna(how='all', axis=1)
df = df.dropna(how='all')

#現代野球ではよっぽどのことが無い限り起こらない成績を削除
df_a = df[df["本塁打"] >= 50]
df_b = df_a[df_a["盗塁"] <= 150]
df_c = df_b[df_b["失策"] <= 100]
df_result = df_c.reset_index()
df_result = df_result.drop("index", axis=1)
df_result
```

```python
df_zyunni = pd.DataFrame(df_result.corr()['順位'])
X_columns = df_zyunni.index.tolist()

# 順位は目的変数のため除外
X_columns.remove('順位')

# 説明変数と目的変数をセット
X = df_result[X_columns]
y = df_result['順位']
```

```python
df_zyunni 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error as MSE
import numpy as np
```

```python
# データの分割
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
 
# モデルの箱の準備および学習
lr = LR()
lr.fit(X_train, y_train)
 
# 予測値の算出
y_pred_test = lr.predict(X_test)
print(y_pred_test)
 
# MSEの算出
mse_test = MSE(y_test, y_pred_test)
 
# RMSEの算出
rmse_test = np.sqrt(mse_test)
 
# RMSEの表示
print(rmse_test)
```

```python
# 予測順位の算出
y_pred_2022 = lr.predict(X_2022)
print(y_pred_2022)
 
# 予測順位列の追加
df_2022['予測順位'] = np.round(y_pred_2022).astype(np.int32)

display(df_2022)
```
